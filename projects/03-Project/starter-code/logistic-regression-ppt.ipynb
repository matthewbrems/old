```python
import pandas as pd
%matplotlib inline
titanic = pd.read_csv('/Users/jennydoyle/Desktop/dsi/self/train.csv')
titanic = pd.get_dummies(titanic, columns=['Embarked','Sex'])
titanic.dropna(axis=0,inplace=True)

X = titanic['Pclass']
y = titanic['Survived']
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from matplotlib import pyplot as plt
import seaborn as sns

# linear = LinearRegression()
# linear.fit(X,y)
# a
# predictions = linear.predict(X)
# plt.plot(predictions, y)
# plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel("Age",fontsize=30)
plt.ylabel("Survived",fontsize=30)
plt.scatter(X,y)
```

```python
sns.set_context("poster")
sns.lmplot(x="Age",y='Survived', data=titanic)
plt.ylim(-2, 2)
plt.xlim(0, 150)
sns.lmplot(x="Age",y='Survived', data=titanic, logistic=True)
plt.xlim(0, 150)
plt.ylim(-2, 2)
```

```python
sns.set_context("poster")
sns.lmplot(x="Pclass",y='Survived', data=titanic)
plt.ylim(-2, 2)
plt.xlim(0, 150)
sns.lmplot(x="Pclass",y='Survived', data=titanic, logistic=True)
plt.xlim(0, 150)
plt.ylim(-2, 2)
```

```python
sns.set_context("talk")
sns.residplot(x="Age",y='Survived', data=titanic)
```

```python

logistic = LogisticRegression()
logistic.fit(X,y)

predictions = logistic.predict(X)
plt.plot(predictions, y)
plt.show()
```
