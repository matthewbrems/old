```python
import pandas as pd
%matplotlib inline
titanic = pd.read_csv('/Users/jennydoyle/Desktop/dsi/self/train.csv')
titanic = pd.get_dummies(titanic, columns=['Embarked','Sex'])
titanic.dropna(axis=0,inplace=True)

X = titanic['Age']
y = titanic['Survived']
```

```python
import statsmodels.formula.api as sm

model = sm.logit(
    "Survived ~ Age",
    data = titanic
).fit()

print model.summary()
```

```python
from sklearn.linear_model import LogisticRegression

# because titanic is a pandas df, we need to reshape X and y
# X_reshaped = X.values.reshape(len(X),1)
# y_reshaped = y.values.reshape(len(y),1)

model = LogisticRegression(C=10000).fit(X,y)

print model.coef_
print model.intercept_
print model.score(X,y)
```

```python





predictions = model.predict_proba(titanic['Age'])
plt.scatter(predictions, y)
plt.show()
```
